#Univariate Analysis (Single Variable)

##Age Distribution

```python
sns.histplot(df['age'], bins=20)
plt.show()
```

##Study Hours

```python
sns.histplot(df['study_hours_per_day'])
plt.show()
```

##Categorical data:

```python
sns.countplot(x='gender', data=df)
plt.show()
```

##Distraction Analysis

Check time spent on distractions.

```python
distractions = ['phone_usage_hours','social_media_hours','youtube_hours','gaming_hours']

df[distractions].mean().sort_values().plot(kind='bar')
plt.show()
```

##Productivity Relationship Analysis

###Study Hours vs Final Grade

```python
sns.scatterplot(x='study_hours_per_day', y='final_grade', data=df)
plt.show()
```

###Sleep Hours vs Productivity


```python
sns.scatterplot(x='sleep_hours', y='productivity_score', data=df)
plt.show()
```

###Correlation Analysis

```python
plt.figure(figsize=(15,9))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.show()
```
